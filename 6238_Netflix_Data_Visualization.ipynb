# UNIVARIATE ANALYSIS

# Q1. Highest Year of releasing Movies and Series? Is There any downfall after that If Yes, Give Reason.

```python
plt.figure(figsize=(10,10))
sns.countplot(y=data['Release_year'], color='#C9A0DC')
plt.yticks(rotation=20)
plt.title("Movie and Series Release Year Count Plot", font="Times New Roman", fontsize=20)
plt.xlabel("Count", font="Times New Roman", fontsize=20, alpha=0.7)
plt.ylabel("Release Year", font="Times New Roman", fontsize=20, alpha=0.7)
plt.tight_layout()
plt.show()
# By looking at below graph we can conclude that most number of movies and series were released in year 2018
# There comes up a sudden downfall in 2019 and 2020 which can be the reason of COVID-19
```

# Q2. Which country creates most number of Movies & Series in Netflix? Also check the less content creator country too.

```python
plt.figure(figsize=(12,10))
plt.title('Top 15 Countries on the basis of Content Creation', font="Times New Roman", fontsize=25)
sns.countplot(data=data,y='Country',order=data['Country'].value_counts().index[0:15],palette='Accent')
plt.xlabel('Count', font="Times New Roman", fontsize=20, alpha=0.7)
plt.ylabel('Countries', font="Times New Roman", fontsize=20, alpha=0.7)
plt.show()
# United States is at Highest rank for creating the most number of series and movies till year 2020.
# Thailand stands for Lowest rank in content creation.
```

# BIVARIATE ANALYSIS

# Q3. Who is the Director with most number of Movies? Also find the least one from Top 10.

```python
movie = data[data['Type']=='Movie']
plt.figure(figsize=(15,8))
plt.title('Directors with Most movies', font="Times New Roman", fontsize=25)
sns.countplot(y='Director',data=movie,order=movie['Director'].value_counts().index[0:10],palette='coolwarm')
plt.xlabel('No. of Movies', font="Times New Roman", fontsize=20, alpha=0.7)
plt.ylabel('Directors', font="Times New Roman", fontsize=20, alpha=0.7)
plt.show()
# The director with most number of movies is Raul Campos, Jan Suter who has made more than 18 Movies.
# The director with least number of movies from Top 10 is S.S.Rajamaouli.
```

# Q4. What is the Average Duration of Movies in Netflix

```python
duration = []
movie = movie[movie['Duration'].notna()]
for i in movie['Duration']:
    duration.append(int(i.strip('min')))
plt.figure(1,figsize=(15,10))
plt.title("Duration of Movies", font="Times New Roman", fontsize=25)
plt.xlabel('Duration', font="Times New Roman", fontsize=20, alpha=0.7)
plt.ylabel('Density', font="Times New Roman", fontsize=20, alpha=0.7)
sns.distplot(duration)
plt.show()
# The average Duration of Movie in Netflix is 90 min
```

# Q5. Show Scatterplot Visualization for Release Year of Movie and Series with Rating of them

```python
plt.figure(figsize=(10, 10))
new = sns.blend_palette(['red','black'],2)
plt.xticks(rotation=90)
sns.scatterplot(x='Release_year', y='Rating',hue="Type", data=data,palette=new)
plt.title("Released Year vs Rating", font="Times New Roman", fontsize=25)
plt.xlabel("Release Year", font="Times New Roman", fontsize=20, alpha=0.7)
plt.ylabel("Ratings", font="Times New Roman", fontsize=20, alpha=0.7)
leg=plt.legend(prop={'family': 'serif', 'size': 10})
leg.set_title("Type of Content", prop={'weight':'bold', 'family': 'serif'})
plt.show()
# By analysing the graph below we can conclude that the rating of explicit content or Parental Guidance Content is more for TV Shows(Web Series)
```
